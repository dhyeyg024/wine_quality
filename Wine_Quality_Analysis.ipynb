## Splitting Dependent and Independent Features 

```python
X=df.drop(['quality'], axis=1)
y=df['quality']
```

## Data Pre-Processing 

### Oversampling the Data 

### As the records having quality equal to 5 & 6 are too much compared to other values of quality, the rows need to made duplicated haveing value of quality equal to 1,2,3,4,7 and 8. 

```python
## oversampling
from imblearn.over_sampling import SMOTE
os=SMOTE()
X_res,y_res=os.fit_resample(X, y)
```

### Train-Test Split 

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_res,y_res,test_size=0.25, random_state=100)
```

### Standardizing the data values 

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler().fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

## Model Building 

### Trying Decision Tree 

```python
from sklearn.tree import DecisionTreeClassifier
tree_model = DecisionTreeClassifier(criterion ='gini')
```

```python
tree_model.fit(X_train,y_train)
tree_model.score(X_test,y_test)
```

### Trying Logistic Regression 

```python
from sklearn.linear_model import LogisticRegression
reg = LogisticRegression()
```

```python
reg.fit(X_train,y_train)
reg.score(X_test,y_test)
```

### Trying Random Forest Classifier 

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=300)
```

```python
model.fit(X_train,y_train)
model.score(X_test,y_test)
```

# Random Forest Classifier gives the best accuracy score. It predicts 86.4% of the data correctly.
